```python
# We only want the 30 that are in Sydney 
# Can create a list that contains these 30 lga's
sydney_lga = ['Bayside (A)','Blacktown (C)','Burwood (A)','Canada Bay (A)','Campbelltown (C) (NSW)','Canterbury-Bankstown (A)',
             'Cumberland (A)','Fairfield (C)','Georges River (A)','Hawkesbury (C)','The Hills Shire (A)','Hornsby (A)',
             'Hunters Hill (A)','Inner West (A)','Ku-ring-gai (A)','Lane Cove (A)','Liverpool (C)','Mosman (A)','North Sydney (A)',
             'Northern Beaches (A)','Parramatta (C)','Penrith (C)','Randwick (C)','Ryde (C)','Strathfield (A)','Sutherland Shire (A)',
             'Waverley (A)','Willoughby (C)', 'Woollahra (A)','Sydney (C)']
```

```python
# We want to subset of data based on values in the lga name column
# Can create new dataframe with only the sydney lga list features in the lga_name19 column

df_sydney = df[df.lga_name19.isin(sydney_lga)]
```

```python
# Less records now because data is only for sydney region
df_sydney.shape
```

```python
df_sydney
```

```python
# Can drop unnecessary columns 
df_sydney = df_sydney.drop(labels=['lhd_2010_code','lhd_2010_name','lga_code19'],axis=1)
df_sydney
```

```python
# Also need to reset index 
df_sydney.reset_index(drop=True, inplace=True)
df_sydney
```
